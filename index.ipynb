## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

SQL databases are relational and house all their data on a single server. The NoSQL databases are generally distributed across an arbitrary number of servers allowing for storage of truly massive repositories of Big Data. This allows for speedier queries as well as redundancy to hedge against servers being down. 

Neither type is necessarily better. It depends on the type(s) of data you are storing and your query and backup needs.





## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A NoSQL database would be a better choice over a relational database if you were dealing with trillions of lines of data (major retailers), if the relationships are complex (social media connections), or if you are storing large numbers of documents that vary in size.

## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Fault tolerance has to do with how the distributed datasets deal with servers being offline for various reasons. With relational databases, if the server is down, you cannot query it at all. With RDD's, each location stores its primary data along with backups for the other locations allowing for analysis even when one or more servers are down.





## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MapReduce is the secret sauce that allows Hadoop to work its magic. It allows for simultaneous queries to multiple sites/servers (map). These servers respond, the responses are aggregated (reduce), and returned to the workstation that initiated the query.
